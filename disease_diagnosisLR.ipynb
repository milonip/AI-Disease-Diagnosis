```python
import pandas as pd

train_df = pd.read_csv('/content/drive/MyDrive/Disease_Diagnosis/training.csv')
test_df = pd.read_csv('/content/drive/MyDrive/Disease_Diagnosis/testing.csv')

train_df.head(), test_df.head()
```

```python
# features and target
X_train = train_df.drop(columns=['prognosis'])
y_train = train_df['prognosis']

X_test = test_df.drop(columns=['prognosis'])
y_test = test_df['prognosis']
```

```python
X_train = X_train.drop(columns=['Unnamed: 133'], errors='ignore')
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)

y_pred = log_model.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
import joblib
joblib.dump(log_model, "disease_diagnosis_model.pkl")
```
